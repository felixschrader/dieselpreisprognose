```python
# Brent Oil Preise von FRED herunterladen und im Notebook-Ordner speichern
import pandas as pd
from pandas_datareader import data as pdr
from datetime import datetime
import os

# 1. FRED API-Key eintragen (kostenlos registrieren: https://fred.stlouisfed.org/docs/api/api_key.html)
FRED_API_KEY = "6e656cd74cb902920c8ae7d3ded9deae"  # <--- Hier deinen API-Key einfügen!

# API-Key für pandas_datareader setzen
os.environ["FRED_API_KEY"] = FRED_API_KEY

# 2. Speicherort: Gleicher Ordner wie das Notebook
csv_pfad = os.path.join(os.getcwd(), "brent_oil_prices.csv")
print(f"Die CSV wird hier gespeichert: {csv_pfad}")

# 3. Funktion zum Herunterladen der Daten
def lade_oelpreise(startdatum="2014-01-01", enddatum=datetime.today().strftime("%Y-%m-%d")):
    try:
        print("Lade Brent Oil Preise von FRED...")
        oelpreise = pdr.DataReader("DCOILBRENTEU", "fred", start=startdatum, end=enddatum)
        oelpreise.to_csv(csv_pfad)
        print(f"✅ Erfolg! Daten gespeichert unter: {csv_pfad}")
        return oelpreise
    except Exception as e:
        print(f"❌ Fehler: {e}")
        return None

# 4. Daten laden und anzeigen
oel_daten = lade_oelpreise()
if oel_daten is not None:
    print("\nVorschau der Daten:")
    display(oel_daten.head())  # Zeigt die ersten 5 Zeilen an
    print(f"\n📊 Datenbereich: {oel_daten.index[0].strftime('%d.%m.%Y')} bis {oel_daten.index[-1].strftime('%d.%m.%Y')}")
    print(f"📄 Dateigröße: {len(oel_daten)} Einträge")


```
